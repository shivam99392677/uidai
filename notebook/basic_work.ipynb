```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
enrollment_data['date'] = pd.to_datetime(enrollment_data['date'],format='%d-%m-%Y')
# Create month names
enrollment_data["month"] = enrollment_data["date"].dt.month_name()
# Define correct month order
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
# Count enrollments per month
month_counts = (
    enrollment_data["month"]
    .value_counts()
    .reindex(month_order)
    .reset_index()
)
month_counts.columns = ["Month", "Number of Enrollments"]

# Plot
fig = px.bar(
    month_counts,
    x="Month",
    y="Number of Enrollments",
    title="Enrollments by Month"
)

fig.show()
```

```python
pin_month_enroll = enrollment_data.groupby(['pincode','month'])[['age_0_5', 'age_5_17', 'age_18_greater']].sum().reset_index()

pin_month_enroll["total_enrollments"] = (
    pin_month_enroll["age_0_5"]
    + pin_month_enroll["age_5_17"]
    + pin_month_enroll["age_18_greater"]
)
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

pin_month_enroll["month"] = pd.Categorical(
    pin_month_enroll["month"],
    categories=month_order,
    ordered=True
)
pin_month_enroll = pin_month_enroll.sort_values(
    ["pincode", "month"]
)
pin_month_enroll["enrollment_velocity"] = (
    pin_month_enroll
        .groupby("pincode")["total_enrollments"]
        .diff()
)
pin_month_enroll['enrollment_velocity'].fillna(0, inplace=True)
pin_month_enroll
```

```python
velocity_by_pincode = (
    pin_month_enroll
        .groupby("pincode")["enrollment_velocity"]
        .mean()
        .reset_index()
)
fig = px.line(
    velocity_by_pincode,
    x="pincode",
    y="enrollment_velocity",
    title="Average Enrollment Velocity by Pincode",
    labels={
        "pincode": "Pincode",
        "enrollment_velocity": "Average Enrollment Velocity"
    }
)
fig.update_layout(
    template="plotly_white",
    xaxis=dict(
        tickangle=-90,
        rangeslider=dict(visible=True)
    )
)

fig.show()

```

## Now I am working on making a anomaly detection.
= for that I needed the lat and long data

```python
# this is the link of open cage to extract the data
import requests
import time
pincodes = enrollment_data['pincode'].unique()
location_data = []
# To respect API rate limits
len(pincodes)
```

```python
link = 'https://api.opencagedata.com/geocode/v1/json?q={pin},India&key=5fb90d83808741528aea82cf6d358774' # 
```

```python

for i, pin in enumerate(pincodes):
    response = requests.get(link.format(pin=pin))

    if response.status_code == 200:
        data = response.json()

        if data.get('results'):
            location_data.append({
                'pincode': pin,
                'latitude': data['results'][0]['geometry']['lat'],
                'longitude': data['results'][0]['geometry']['lng']
            })
        else:
            location_data.append({
                'pincode': pin,
                'latitude': None,
                'longitude': None
            })

    time.sleep(1)

print(i)
```

```python
# Create a dataframe from location_data
location_df = pd.DataFrame(location_data)

# Save it to a CSV file
location_df.to_csv(r'C:\Users\ankit\Desktop\uidai\data_set\location_data.csv', index=False)

print('Location dataframe created and saved')
print(location_df.head())
```

```python
from sklearn.cluster import DBSCAN

coords = enrollment_data[['latitude', 'longitude']].to_numpy()
db = DBSCAN(eps=0.1, min_samples=10).fit(coords)
enrollment_data['cluster'] = db.labels_
import plotly.express as px

fig = px.scatter_mapbox(
    enrollment_data,
    lat="latitude",
    lon="longitude",
    color="cluster",
    zoom=4,
    mapbox_style="carto-positron"
)

fig.show()
```
